Fill each missing nominal feature value with the string “Missing”.

```python
filled_housing = df_housing.fillna('missing')
filled_housing.head()
```

Interpolate each missing numeric feature value using linear interpolation.

```python
filled_linear = filled_housing.select_dtypes(exclude=["object"]).interpolate(method='linear')
filled_linear = filled_linear.drop(columns=['Order','PID'])
filled_linear.head()
```

We can see in this heatmap that we have no more missing values in the dataframe.

```python
sns.heatmap(filled_linear.isnull(),yticklabels=False,cbar=False)
```

2.2 Standardization

Standardize the imputed feature data so that the values of each numeric feature are standard normally distributed (i.e., each feature is Gaussian with zero mean and unit variance).

```python
filled_linear_standard = filled_linear.transform(lambda x: (x - x.mean()) / x.std(ddof=1))
filled_linear_standard.head()
```

Visualize the results using box plots. How do the plots differ from box plots made before feature standardization? Which feature has the outlier furthest from the mean before and after standardization?

Answer: The Sale Price feature contains the outliers that are furthest from the mean before standardization, and the Misc Val feature has the furthest outlier after standardization. 

```python
plt.figure(figsize=(10,6))
ax = sns.boxplot(data=filled_linear)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.title("Before Feature Standardization")
plt.show()
```

filled_linear_standard is the standardized feature data

```python
plt.figure(figsize=(10,6))
ax = sns.boxplot(data=filled_linear_standard)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.title("After Feature Standardization")
plt.show()
```

2.3 Feature Selection

To get an idea of their relative importance, estimate the mutual information be- tween the numeric features and the class column, ‘SalePrice’. Note that scikit-learn may be used to estimate mutual information (see sklearn.feature selection).

```python
X = filled_linear_standard.iloc[:, :-1]
y = filled_linear_standard.iloc[:,-1]
```

```python
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import SelectKBest
```

```python
mi = mutual_info_regression(X, y, random_state=None)
print(mi)
```

```python
# Select the k best features.
selector = SelectKBest(mutual_info_regression, k=5)
selector.fit(X, y)

# Get indices of columns to keep.
idxs_selected = selector.get_support(indices=True)

# Create new DataFrame with only desired columns.
X_new = X.iloc[:,idxs_selected]

# Show the top 5 features.
#idxs_selected
print(X_new.head())
```

```python
info_columns = X_new.columns
info_columns
```

```python
info_values = mi[idxs_selected]
info_values
```

What are the top 5 numeric features ranked by mutual information? Note that features with a higher estimated mutual information are considered more informative.

```python
info_list = dict(zip(info_columns, info_values))
info_list
```

How do you expect the values for the top-ranked feature to affect the sales price (i.e., would you expect the sales price to increase when its values go up or down)? Why?

Answer: 

In this dataset, we measured the dependency of the Sale Price on the top_ranked features through mutual information regression, which estimates mutual information for the continuous target variable (Sale Price).

These top_ranked features (Overall Qual, Year Built, Year Remod/Add, 1st Flr SF, Gr Liv Area), had the highest values which means that the Sale Price was most dependent on these features.

We can expect the Sale Price of a home to be dependent on the top_features. For instance, a low Overall Qual in a home would most likely bring the value of the Sale Price down; whereas, a house with a high Overall Qual feature would probably sell for more. 

Other features such as, Year Built, Year Remod/Add, 1st Flr SF, and Gr Liv Area, would also have a significant impact on the Sale Price.

In this next section, I extended this assignment by creating a linear regression model. What I found in this data was quite astonishing because MOST of the top_features in the linear regression model coefficients matched the top features in the mutual information regression selector. 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

Top_Features In Linear Regression Model Coefficients

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df.sort_values(by='Coefficient', ascending=False).head()
```

```python
predictions = lm.predict(X_test)
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of Residuals")
plt.scatter(y_test,predictions)
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of Residuals")
sns.distplot(y_test-predictions);
```

```python
from sklearn import metrics
import numpy as np
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
